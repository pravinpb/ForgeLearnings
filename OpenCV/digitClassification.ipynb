```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
train_images = train_images.reshape((train_images.shape[0], 28 * 28))
test_images = test_images.reshape((test_images.shape[0], 28 * 28))
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(28 * 28,)),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=10, batch_size=32)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(loss=tf.keras.losses.categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,
          batch_size=128,
          epochs=20,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def process_frame(frame):
    # Convert the frame to grayscale
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Resize the frame to the expected input size of the model
    frame = cv2.resize(frame, (28, 28))  # Adjust size according to your model's input
    
    # Normalize the frame
    frame = frame.astype('float32') / 255
    
    # Expand dimensions to match the model's input shape
    frame = np.expand_dims(frame, axis=0)
    
    # Perform classification
    prediction = model.predict(frame)
    
    # Get the predicted class
    predicted_class = np.argmax(prediction)
    
    return predicted_class
```

```python
# Open the video file
cap = cv2.VideoCapture('/Users/pravinpb/pycode/openCV/workshop/assets/MNIST Dream video.mp4')

while True:
    # Read the next frame from the video
    ret, frame = cap.read()
    
    if not ret:
        break  # Exit the loop if the frame is empty
    
    # Process the frame
    predicted_class = process_frame(frame)
    
    # Display the frame with the prediction
    cv2.putText(frame, f'Predicted Class: {predicted_class}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Display the frame
    cv2.imshow('Frame', frame)
    
    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close windows
cap.release()
cv2.destroyAllWindows()
```
